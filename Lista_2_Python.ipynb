```python
# UNINASSAU
# DOCENTE: CLOVES ROCHA
# ESTUDANTE: DANILO MIGUEL

# Obter as notas de um aluno correspondente com base na média
def obter_nota(media):
    if 90 <= media <= 100:
        return "A"
    elif 80 <= media < 90:
        return "B"
    elif 70 <= media < 80:
        return "C"
    elif 60 <= media < 70:
        return "D"
    else:
        return "F"

# Solicitar ao usuário a quantidade de notas
num_notas = int(input("Quantas notas você deseja inserir? "))

# Inicializar a soma das notas
soma_notas = 0

# Solicitar ao usuário as notas e somá-las
for i in range(num_notas):
    nota = float(input(f"Insira a nota {i + 1}: "))
    soma_notas += nota

# Calcular a média
media = soma_notas / num_notas

# Obter a nota correspondente
nota_correspondente = obter_nota(media)

# Exibir a média e a nota correspondente
print(f"A média das notas é: {media}")
print(f"A nota correspondente é: {nota_correspondente}")
```

```python
import csv
import os

# Função para criar o arquivo CSV se ele não existir
def criar_agenda(arquivo):
    if not os.path.exists(arquivo):
        with open(arquivo, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["Nome", "Telefone"])

# Função para adicionar um contato à agenda
def adicionar_contato(arquivo, nome, telefone):
    with open(arquivo, 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([nome, telefone])
    print(f"Contato {nome} adicionado com sucesso!")

# Função para listar todos os contatos na agenda
def listar_contatos(arquivo):
    with open(arquivo, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Ignora o cabeçalho
        for row in reader:
            print(f"Nome: {row[0]}, Telefone: {row[1]}")

# Função para buscar um contato na agenda
def buscar_contato(arquivo, nome):
    with open(arquivo, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Ignora o cabeçalho
        for row in reader:
            if row[0].lower() == nome.lower():
                print(f"Nome: {row[0]}, Telefone: {row[1]}")
                return
        print(f"Contato {nome} não encontrado na agenda.")

# Função para remover um contato da agenda
def remover_contato(arquivo, nome):
    contatos = []
    with open(arquivo, 'r') as file:
        reader = csv.reader(file)
        header = next(reader)
        for row in reader:
            if row[0].lower() != nome.lower():
                contatos.append(row)
    with open(arquivo, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(header)
        writer.writerows(contatos)
    print(f"Contato {nome} removido com sucesso!")

if __name__ == "__main__":
    arquivo_csv = "agenda.csv"
    criar_agenda(arquivo_csv)

    while True:
        print("\nOpções da Agenda:")
        print("1. Adicionar Contato")
        print("2. Listar Contatos")
        print("3. Buscar Contato")
        print("4. Remover Contato")
        print("5. Sair")

        escolha = input("Escolha uma opção (1/2/3/4/5): ")

        if escolha == "1":
            nome = input("Digite o nome do contato: ")
            telefone = input("Digite o telefone do contato: ")
            adicionar_contato(arquivo_csv, nome, telefone)
        elif escolha == "2":
            print("\nLista de Contatos:")
            listar_contatos(arquivo_csv)
        elif escolha == "3":
            nome = input("Digite o nome do contato que deseja buscar: ")
            buscar_contato(arquivo_csv, nome)
        elif escolha == "4":
            nome = input("Digite o nome do contato que deseja remover: ")
            remover_contato(arquivo_csv, nome)
        elif escolha == "5":
            print("Saindo da Agenda.")
            break
        else:
            print("Opção inválida. Tente novamente.")
```

```python
import random

# Lista de palavras para o jogo
palavras = ["elefante", "abacaxi", "universo", "sereia", "guitarra", "futebol",
           "chocolate", "esquilo", "arco-íris", "astronauta", "xadrez", "girassol",
           "montanha", "pinguim", "biblioteca", "espaço", "morango", "hipopótamo",
           "cachoeira", "palhaço"]

# Função para escolher uma palavra aleatória da lista
def escolher_palavra():
    return random.choice(palavras)

# Função para mostrar a palavra oculta com letras adivinhadas
def mostrar_palavra(palavra, letras_adivinhadas):
    resultado = ""
    for letra in palavra:
        if letra in letras_adivinhadas:
            resultado += letra
        else:
            resultado += "_"
    return resultado

# Função principal do jogo
def jogo_da_forca():
    palavra = escolher_palavra()
    letras_adivinhadas = []
    tentativas = 6  # Número de tentativas permitidas

    print("Bem-vindo ao Jogo da Forca!")

    while tentativas > 0:
        palavra_oculta = mostrar_palavra(palavra, letras_adivinhadas)
        print(f"Palavra: {palavra_oculta}")
        print(f"Tentativas restantes: {tentativas}")

        try:
            palpite = input("Digite uma letra: ").lower()
        except KeyboardInterrupt:
            print("Jogo interrompido pelo jogador.")
            break  # Sair do loop do jogo

        if len(palpite) != 1 or not palpite.isalpha():
            print("Por favor, insira uma única letra válida.")
            continue

        if palpite in letras_adivinhadas:
            print(f"Você já tentou a letra {palpite}.")
            continue

        letras_adivinhadas.append(palpite)

        if palpite in palavra:
            print("Letra correta!")
            if palavra_oculta == palavra:
                print("Parabéns, você venceu! A palavra é:", palavra)
                break
        else:
            print("Letra incorreta.")
            tentativas -= 1

    if tentativas == 0:
        print(f"Você perdeu! A palavra correta era: {palavra}")

if __name__ == "__main__":
    jogo_da_forca()
```

```python
class ContaBancaria:
    def __init__(self, titular, saldo=0):
        self.titular = titular
        self.saldo = saldo

    def consultar_saldo(self):
        return self.saldo

    def depositar(self, valor):
        if valor > 0:
            self.saldo += valor
            return f"Depósito de R${valor} realizado com sucesso. Novo saldo: R${self.saldo}"
        else:
            return "Valor de depósito inválido."

    def sacar(self, valor):
        if valor > 0 and valor <= self.saldo:
            self.saldo -= valor
            return f"Saque de R${valor} realizado com sucesso. Novo saldo: R${self.saldo}"
        elif valor <= 0:
            return "Valor de saque inválido."
        else:
            return "Saldo insuficiente"

    def alterar_titular(self, novo_titular):
        self.titular = novo_titular
        return f"Titular da conta alterado para {novo_titular}"

def main():
    print("Bem-vindo ao Simulador de Caixa Eletrônico")
    nome_titular = input("Digite o nome do titular da conta: ")
    conta = ContaBancaria(nome_titular)

    while True:
        print("\nEscolha uma opção:")
        print("1. Consultar Saldo")
        print("2. Depositar")
        print("3. Sacar")
        print("4. Alterar Titular da Conta")
        print("5. Sair")

        opcao = input("Digite o número da opção desejada: ")

        if opcao == "1":
            saldo = conta.consultar_saldo()
            print(f"Saldo disponível: R${saldo}")
        elif opcao == "2":
            valor = float(input("Digite o valor a ser depositado: R$"))
            mensagem = conta.depositar(valor)
            print(mensagem)
        elif opcao == "3":
            valor = float(input("Digite o valor a ser sacado: R$"))
            mensagem = conta.sacar(valor)
            print(mensagem)
        elif opcao == "4":
            novo_titular = input("Digite o novo titular da conta: ")
            mensagem = conta.alterar_titular(novo_titular)
            print(mensagem)
        elif opcao == "5":
            print("Saindo do Simulador de Caixa Eletrônico. Obrigado!")
            break
        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
    main()
```

```python
# Dicionário de taxas de câmbio (taxa para 1 unidade da moeda base)
taxas_de_cambio = {
    'USD': 1.0,
    'EUR': 0.85,
    'GBP': 0.74,
    'JPY': 110.17,
}

def converter_moeda(valor, moeda_origem, moeda_destino):
    if moeda_origem in taxas_de_cambio and moeda_destino in taxas_de_cambio:
        taxa_origem = taxas_de_cambio[moeda_origem]
        taxa_destino = taxas_de_cambio[moeda_destino]
        valor_convertido = valor * (taxa_destino / taxa_origem)
        return valor_convertido
    else:
        return None

def main():
    print("Bem-vindo ao Conversor de Moedas")
    moeda_origem = input("Digite a moeda de origem (USD, EUR, GBP, JPY): ").upper()
    moeda_destino = input("Digite a moeda de destino (USD, EUR, GBP, JPY): ").upper()
    valor = float(input("Digite o valor a ser convertido: "))

    resultado = converter_moeda(valor, moeda_origem, moeda_destino)

    if resultado is not None:
        print(f"{valor} {moeda_origem} equivalem a {resultado} {moeda_destino}")
    else:
        print("Moeda de origem ou de destino não suportada.")

if __name__ == "__main__":
    main()
```

```python
import json

# Função para carregar tarefas do arquivo JSON
def carregar_tarefas():
    try:
        with open("tarefas.json", "r") as arquivo:
            tarefas = json.load(arquivo)
    except FileNotFoundError:
        tarefas = []
    return tarefas

# Função para salvar tarefas no arquivo JSON
def salvar_tarefas(tarefas):
    with open("tarefas.json", "w") as arquivo:
        json.dump(tarefas, arquivo, indent=4)

# Função para adicionar uma nova tarefa
def adicionar_tarefa(tarefas, descricao):
    tarefa = {"descricao": descricao, "concluida": False}
    tarefas.append(tarefa)
    print(f"Tarefa '{descricao}' adicionada com sucesso!")

# Função para listar as tarefas
def listar_tarefas(tarefas):
    if not tarefas:
        print("Nenhuma tarefa encontrada.")
    else:
        for i, tarefa in enumerate(tarefas, start=1):
            status = "Concluída" if tarefa["concluida"] else "Pendente"
            print(f"{i}. [{status}] {tarefa['descricao']}")

# Função para marcar uma tarefa como concluída
def marcar_tarefa_concluida(tarefas, numero_tarefa):
    if 1 <= numero_tarefa <= len(tarefas):
        tarefa = tarefas[numero_tarefa - 1]
        tarefa["concluida"] = True
        print(f"Tarefa '{tarefa['descricao']}' marcada como concluída.")
    else:
        print("Número de tarefa inválido.")

# Função para remover uma tarefa
def remover_tarefa(tarefas, numero_tarefa):
    if 1 <= numero_tarefa <= len(tarefas):
        tarefa = tarefas.pop(numero_tarefa - 1)
        print(f"Tarefa '{tarefa['descricao']}' removida com sucesso.")
    else:
        print("Número de tarefa inválido.")

def main():
    tarefas = carregar_tarefas()

    while True:
        print("\nOpções do Gerenciador de Tarefas:")
        print("1. Adicionar Tarefa")
        print("2. Listar Tarefas")
        print("3. Marcar Tarefa como Concluída")
        print("4. Remover Tarefa")
        print("5. Sair")

        opcao = input("Escolha uma opção (1/2/3/4/5): ")

        if opcao == "1":
            descricao = input("Digite a descrição da tarefa: ")
            adicionar_tarefa(tarefas, descricao)
        elif opcao == "2":
            print("\nLista de Tarefas:")
            listar_tarefas(tarefas)
        elif opcao == "3":
            listar_tarefas(tarefas)
            numero_tarefa = int(input("Digite o número da tarefa concluída: "))
            marcar_tarefa_concluida(tarefas, numero_tarefa)
        elif opcao == "4":
            listar_tarefas(tarefas)
            numero_tarefa = int(input("Digite o número da tarefa a ser removida: "))
            remover_tarefa(tarefas, numero_tarefa)
        elif opcao == "5":
            salvar_tarefas(tarefas)
            print("Saindo do Gerenciador de Tarefas. Tarefas salvas.")
            break
        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
    main()
```

```python
import random

# Gerar um número aleatório entre 1 e 100
numero_secreto = random.randint(1, 100)
tentativas = 0

print("Bem-vindo ao Jogo de Adivinhação!")
print("Tente adivinhar o número entre 1 e 100.")

while True:
    tentativa = input("Digite um número: ")
    try:
        tentativa = int(tentativa)
    except ValueError:
        print("Por favor, insira um número válido.")
        continue

    tentativas += 1

    if tentativa < numero_secreto:
        print("Tente um número maior.")
    elif tentativa > numero_secreto:
        print("Tente um número menor.")
    else:
        print(f"Parabéns! Você adivinhou o número {numero_secreto} em {tentativas} tentativas.")
        break
```

```python
# Função para calcular o IMC
def calcular_imc(peso, altura):
    return peso / (altura ** 2)

# Função para determinar a categoria de IMC
def categoria_imc(imc):
    if imc < 18.5:
        return "Abaixo do peso"
    elif 18.5 <= imc < 24.9:
        return "Peso saudável"
    elif 24.9 <= imc < 29.9:
        return "Sobrepeso"
    else:
        return "Obesidade"

# Solicitar peso e altura ao usuário
peso = float(input("Digite o seu peso em quilogramas: "))
altura = float(input("Digite a sua altura em metros: "))

# Calcular o IMC
imc = calcular_imc(peso, altura)

# Determinar a categoria de IMC
categoria = categoria_imc(imc)

# Exibir o resultado
print(f"Seu IMC é {imc:.2f} e você está na categoria: {categoria}")
```

```python
# Função para mostrar a lista de filmes
def mostrar_filmes():
    print("Filmes disponíveis:")
    for i, filme in enumerate(filmes, start=1):
        print(f"{i}. {filme}")

# Função para mostrar a lista de horários
def mostrar_horarios():
    print("Horários disponíveis:")
    for i, horario in enumerate(horarios, start=1):
        print(f"{i}. {horario}")

# Função para fazer uma reserva
def fazer_reserva(filme, horario, quantidade):
    with open("reservas.txt", "a") as arquivo:
        arquivo.write(f"Filme: {filmes[filme-1]}, Horário: {horarios[horario-1]}, Ingressos: {quantidade}\n")

filmes = ["Filme 1", "Filme 2", "Filme 3"]
horarios = ["10:00", "14:30", "18:00"]

print("Bem-vindo ao Sistema de Reservas de Cinema!")

while True:
    mostrar_filmes()
    escolha_filme = int(input("Escolha um filme (1/2/3) ou 0 para sair: "))

    if escolha_filme == 0:
        print("Saindo do sistema de reservas. Obrigado!")
        break
    elif 1 <= escolha_filme <= len(filmes):
        mostrar_horarios()
        escolha_horario = int(input("Escolha um horário (1/2/3): "))

        if 1 <= escolha_horario <= len(horarios):
            quantidade = int(input("Quantos ingressos deseja comprar? "))
            fazer_reserva(escolha_filme, escolha_horario, quantidade)
            print("Reserva realizada com sucesso!")
        else:
            print("Opção de horário inválida. Tente novamente.")
    else:
        print("Opção de filme inválida. Tente novamente.")
```

```python
import requests

# Chave de API da OpenWeatherMap
api_key = "SUA_CHAVE_DE_API_AQUI"  # Substitua pela sua chave

def obter_previsao_tempo(cidade):
    base_url = "https://api.openweathermap.org/data/2.5/weather"
    parametros = {
        "q": cidade,
        "appid": api_key,
        "units": "metric"  # Unidades em Celsius
    }

    resposta = requests.get(base_url, params=parametros)

    if resposta.status_code == 200:
        dados = resposta.json()
        temperatura = dados["main"]["temp"]
        descricao = dados["weather"][0]["description"]
        print(f"Previsão do tempo para {cidade}:")
        print(f"Temperatura: {temperatura}°C")
        print(f"Descrição: {descricao}")
    else:
        print("Não foi possível obter a previsão do tempo. Verifique o nome da cidade e a chave de API.")

def main():
    print("Aplicativo de Previsão do Tempo")
    cidade = input("Digite o nome da cidade: ")
    obter_previsao_tempo(cidade)

if __name__ == "__main__":
    main()
```
